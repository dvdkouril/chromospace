```python
# Downloading the sample Parquet file
import urllib.request

url = "https://raw.githubusercontent.com/dvdkouril/chromospace-sample-data/main/chr11.parquet"
with urllib.request.urlopen(url) as f:
    res = f.read()
    f = open("chr11.parquet", "wb")
    f.write(res)
    f.close()
```

```python
# Reading the Parquet file to an Arrow table
import pyarrow.parquet as pq

table = pq.read_table("chr11.parquet")
print(table)
```

```python
# Convert Arrow Table to bytes
output_stream = pa.BufferOutputStream()
with pa.RecordBatchStreamWriter(output_stream, table.schema) as writer:
    writer.write_table(table)

# Get the bytes
arrow_bytes = output_stream.getvalue().to_pybytes()

print(arrow_bytes)
```
